```
#1
#Factorial of a number
a=int(input("Enter a Number\n"))
fact=1
if (a < 0):
  print("Factorials does not exist for negative number")
elif (a == 0):
  print("Factorial of 0 is 1")
else:
  for i in range(1,a + 1):
    fact=fact*i
  print("Factorial of",a,"is",fact)
```

```
#2
#Fibonacci sequence
nterms = int(input("How many terms you want? "))  
# first two terms  
n1 = 0  
n2 = 1  
count = 2  
# check if the number of terms is valid  
if nterms <= 0:  
   print("Plese enter a positive integer")  
elif nterms == 1:  
   print("Fibonacci sequence:")  
   print(n1)  
else:  
   print("Fibonacci sequence:")  
   print(n1,",",n2,end=', ')  
   while count < nterms:  
       nth = n1 + n2  
       print(nth,end=' , ')  
       # update values  
       n1 = n2  
       n2 = nth  
       count += 1
```

```
#5
#Display pyramid with step number accepted from user 
row=int(input("Enter how many lines: \n"))
for i in range(1,row+1):
  for j in range(1,row+1-i):
      print(' ',end='')
  for j in range(1,i+1):
       print(j, end='')
  for j in range(i-1,0,-1):
       print(j, end='')
  print()
```

```
#6
#Count no of characters in a string
def char_frequency(str1):
    dict = {}
    for n in str1:
        keys = dict.keys()
        if n in keys:
            dict[n] += 1
        else:
            dict[n] = 1
    return dict
print(char_frequency('Welcome to ooty nice to meet you'))
```

```
#7
#Add 'ing' at the end of the given string.if it already ends with 'ing' then add 'ly'
def add_string(str1):
  length = len(str1)
  if length > 1:
    if str1[-3:] == 'ing':
     str1 += 'ly'
    else:
     str1 += 'ing'
  return str1
print(add_string('vb'))
print(add_string('webprogramming'))
print(add_string('java'))     
```

```
#8
#Accept a list of words and retutn length of longest word
def longestLength(a): 
    max1 = len(a[0]) 
    temp = a[0] 
  
    # for loop to traverse the list 
    for i in a: 
        if(len(i) > max1): 
  
            max1 = len(i) 
            temp = i 
  
    print("The word with the longest length is:", temp, 
          " and length is ", max1) 
  
  
# Driver Program 
a = input("Enter a list elements separated by space ")

a = a.split() 
longestLength(a) 
```

```
#9
#Construct the following patterns using nested loop 
size=int(input("How many lines you want?: \n"))
m = (2 * size) - 2
for i in range(0, size):
    for j in range(0, m):
        print(end=" ")
    m = m - 1  
    for j in range(0, i + 1):
        print("* ", end=' ')
    print(" ")
```

```
#10
#Generate all factors of a number
x=int(input("Enter a number:\n"))
def print_factors(x):
   print("The factors of",x,"are:")
   for i in range(1, x + 1):
       if x % i == 0:
           print(i)
print_factors(x)
```

```
#11
#Lambda function to  find area of square,rectangle and triangle
import math
t_peri =lambda p,q,r :p+q+r
r_area=lambda len,ht :len*ht 
c_peri=lambda rad :2*math.pi*rad 
c_area=lambda rad :math.pi*rad*rad 

print("perimeter of triangle (10,30,15) is ",t_peri(10,20,15))
print("area of rectangle (30,10) is",r_area(30,20))
print("perimeter of circle (5) is:",c_peri(10))
print("area of circle (20) is: ",c_area(15))
```
